```python
total_installs = df1.groupby(['Genres']).Installs.sum()  # Find the total for each "Installs" unique value.
total_installs.head()
```

```python
mean_installs = df1.groupby(['Genres']).Installs.mean()  # Find the mean for each "Installs" unique value.
mean_installs
```

```python
myPlot = total_installs.plot(kind='bar')    # determine if a bar plot works for total installs.
```

```python
myPlot = mean_installs.plot(kind='bar')    # determine if a bar plot works for mean installs.
```

```python
#  Plot the data in descending order of total app installations:

my_plot = total_installs.sort_values(ascending=False).plot(kind='bar',legend=None,title="App Genre vs. Total Installs", fontsize=10, figsize = (21, 6))
my_plot.set_xlabel("App Genre", fontsize=20)
my_plot.set_ylabel("Total Installs (x10 billion)", fontsize=20);
```

```python
#  Plot the data alphabetically of the total app installs:

my_plot = total_installs.sort_index(ascending=True).plot(kind='bar',legend=None,title="App Genre vs. Total Installs (Alphabetical Order)", fontsize=10, figsize = (21, 6))
my_plot.set_xlabel("App Genre", fontsize=20)
my_plot.set_ylabel("Total Installs (x10 billion)", fontsize=20);
```

```python
#  Plot the data in descending order of mean app installations:

my_plot = mean_installs.sort_values(ascending=False).plot(kind='bar',legend=None,title="App Genre vs. Mean Installs", fontsize=10, figsize = (21, 6))
my_plot.set_xlabel("App Genre", fontsize=20)
my_plot.set_ylabel("Mean Installs (x10 million)", fontsize=20);

```

```python
#  Plot the data alphabetically of the mean app installs:

my_plot = mean_installs.sort_index(ascending=True).plot(kind='bar',legend=None,title="App Genre vs. Mean Installs (Alphabetical Order)", fontsize=10, figsize = (21, 6))
my_plot.set_xlabel("App Genre", fontsize=20)
my_plot.set_ylabel("Mean Installs (x10 million)", fontsize=20);
```
