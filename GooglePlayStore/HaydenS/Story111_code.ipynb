```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.options.mode.chained_assignment = None  # default='warn'
```

```python
data = pd.read_csv('cleaned_googleplaystore.csv')
data.head()
```

```python
free = data[data.Type == 'Free']
free.head()
```

```python
paid = data[data.Type == 'Paid']
paid.head()
```

```python
#mean price for apps with more than 10,000+ installs

#to directly answer the user story:
#horizontal bar chart of only paid apps - divided by category

#violin or box? plot of paid apps - category


#demonstrate impact of including free apps:

#frequency chart of paid apps
#frequency chart of free apps
#side by side bar chart of paid and paid + free apps - divided by category
```

```python
free_install = free[free.Installs >= 10000]
free_install.head()
```

```python
paid_install = paid[paid.Installs >= 10000]
paid_install.head()
```

```python
paid_install.Installs.unique()
```

```python
install_unique = paid_install.Installs.unique()

temp_dict = {}
for i in range(len(install_unique)):
    temp_row = {install_unique[i]:[len(paid_install[paid_install.Installs == install_unique[i]])]}
    temp_dict.update(temp_row)

df = pd.DataFrame(temp_dict)  
df = df.melt()
df.columns = ['installs', 'frequency']
df = df.sort_values(by='installs', ascending=True)
df = df.reset_index(drop=True)

df
```

```python
df.plot.bar(x='installs', y='frequency', rot=0, figsize=(14,7))
```

```python
install_unique = paid_install.Installs.unique()

temp_dict = {}
for i in range(len(install_unique)):
    temp_var = np.mean(paid_install.Price[paid_install.Installs == install_unique[i]])
    temp_var = np.round_(temp_var, decimals=2)
    temp_row = {install_unique[i]:[temp_var]}
    temp_dict.update(temp_row)

df = pd.DataFrame(temp_dict)  
df = df.melt()
df.columns = ['installs', 'price']
df = df.sort_values(by='installs', ascending=True)
df = df.reset_index(drop=True)

df
```

```python
df.plot.bar(x='installs', y='price', rot=0, figsize=(14,7))
```
